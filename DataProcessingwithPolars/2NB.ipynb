```python
import polars as pl
import numpy as np

df = pl.DataFrame({
    "name": ["ALice","bob", "Charlie"],
    "age":[25, 30 , 73],
    "city": ["Nyc","LA", "Chicago"]
})

print(df)
```

```python
print(df.head)
print(df.tail)
print(df.shape)
print(df.columns)
print(df.dtypes)

print(df.describe)
print(df.null_count())
```

```python
result = df.select(["name" , "age"])
print(result)
result1 = df.select(pl.col("name") , pl.col("age"))
print(result1)

```

```python
fil = df.filter(pl.col("age") > 25)
fil1 = df.filter((pl.col("age")>=25) & (pl.col("city") == "Nyc"))
print(fil)
print(fil1)
```

```python
#select and filter togerther
res = df.select(["name" , "age"]).filter(pl.col("age") > 25)
print(res)
```

```python
new = df.with_columns([
    (pl.col("age") + 1).alias("age_next_year"),
    (pl.when(pl.col("age") > 30)
     .then(pl.lit("old"))
        .otherwise(pl.lit("young"))
        .alias("age_group"))
])

print(new)

new2 = df.with_columns([
    pl.col("name").str.to_uppercase().alias("name_upper"),
    pl.col("city").str.len_chars().alias("city_name_length")
])

print(new2)
```

```python
gro = df.group_by("city").agg([
    pl.col("age").mean().alias("avg_age"),
    pl.col("age").max().alias("max_age"),
    pl.count().alias("count")
])
print(gro)

gro2 = df.group_by(["city"]).agg([
    pl.col("age").mean(),
    pl.count()
])
print(gro2)
```

```python
sor = df.sort("age")
sor1 = df.sort(["age"], descending=[True])
sor2 = df.sort(by=["city", "age"])

print(sor)
print(sor1)
print(sor2)

result = df.with_columns([
    pl.col("age").rank().alias("age_rank")
])
print(result)
```

```python
df2 = pl.DataFrame({
    "name": ["ALice", "bob", "David"],
    "salary": [70000, 80000, 120000]
})

result = df.join(df2, on="name", how="inner")
print(result)

left = df.join(df2, on="name", how="left")
print(left)

right = df.join(df2, on="name", how="right")    
print(right)

multi = df.join(df2, on=["name", "salary"], how="inner")
```

```python
result = df.with_columns([
  
    pl.col("age").sum().over("city").alias("total_age_by_city"),
    pl.col("age").mean().over("city").alias("avg_age_by_city"),
    pl.col("age").max().over("city").alias("max_age_by_city")  
])

print(result)
```
